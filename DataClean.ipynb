```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("input.csv")
```

```python
data.head()
```

```python
data1 = pd.DataFrame()
```

```python
data1['A1'], data1['A1_ratio'] = data['A_1'].str.split('/',1).str
data1['A2'], data1['A2_ratio'] = data['A_2'].str.split('/',1).str
data1['A3'], data1['A3_ratio'] = data['A_3'].str.split('/',1).str
data1['B1'], data1['B1_ratio'] = data['B_1'].str.split('/',1).str
data1['B2'], data1['B2_ratio'] = data['B_2'].str.split('/',1).str
data1['X1'], data1['X1_ratio'] = data['X_1'].str.split('/',1).str
data1['X2'], data1['X2_ratio'] = data['X_2'].str.split('/',1).str
data1['PCE'] =data['Max PCE']
```

```python

data1['A1_ratio'].convert_objects(convert_numeric=True)
data1['A2_ratio'].convert_objects(convert_numeric=True)
data1['A3_ratio'].convert_objects(convert_numeric=True)
data1['B1_ratio'].convert_objects(convert_numeric=True)
data1['B2_ratio'].convert_objects(convert_numeric=True)
data1['X1_ratio'].convert_objects(convert_numeric=True)
data1['X2_ratio'].convert_objects(convert_numeric=True)
```

```python
data1 = data1.fillna(0)
```

```python
data1
```

```python
data1['A_sum'] = data1['A1_ratio'].convert_objects(convert_numeric=True)+data1['A2_ratio'].convert_objects(convert_numeric=True)+data1['A3_ratio'].convert_objects(convert_numeric=True)
data1['B_sum'] = data1['B1_ratio'].convert_objects(convert_numeric=True)+data1['B2_ratio'].convert_objects(convert_numeric=True)
data1['X_sum'] = data1['X1_ratio'].convert_objects(convert_numeric=True)+data1['X2_ratio'].convert_objects(convert_numeric=True)
```
